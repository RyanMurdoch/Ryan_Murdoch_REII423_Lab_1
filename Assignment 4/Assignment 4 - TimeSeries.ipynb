```python
import pandas as pd
import datetime
import csv
import matplotlib.pyplot as plt

filename = "CarSales.csv"
f = open(filename,'r')
df = pd.read_csv(f)

df1 = pd.to_datetime(df['Date'], format = '%Y')

df2 = df.set_index('Date')

fig = plt.gcf()
fig.set_size_inches(15, 10)
plt.xlabel('Year')
plt.ylabel('Car Sales')
plt.title('BMW Car Sales')
plt.plot(df2)
plt.savefig('CarSalesTimeSeries.jpg', dpi = 1200)
plt.show()
```
