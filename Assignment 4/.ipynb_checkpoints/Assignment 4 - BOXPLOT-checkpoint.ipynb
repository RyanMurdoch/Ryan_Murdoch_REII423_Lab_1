```python
#Box plots are used to show the shape of the distribution, its central value, and its variability.

import pandas as pd

import matplotlib.pyplot as plt

data = pd.read_csv('brain_size.csv', sep=',', na_values='.')
gender_data = data.groupby('Gender')
gender_data.boxplot(column=['FSIQ', 'VIQ', 'PIQ'])
plt.savefig('BrainBoxPlot.jpg', dpi = 1200)
plt.show()
```
