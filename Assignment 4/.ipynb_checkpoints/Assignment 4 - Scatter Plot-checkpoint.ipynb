```python
#Scatter plots show how much one variable is ralated to another, thus their correlation.
#It can be seen that people with a weight of around 90 has a height of around 1.7m in this dataset

import pandas

import matplotlib.pyplot as plt

data = pandas.read_csv('brain_size.csv', sep=',', na_values='.')
gender_data = data.groupby('Gender')

figure1 = plt.scatter(data['Weight'], data['Height'])
plt.xlabel('Weight (kg)')
plt.ylabel('Height (cm)')
plt.title('Weight vs. Height')
plt.savefig('BrainScatter.jpg', dpi = 1200)
plt.show()
```
