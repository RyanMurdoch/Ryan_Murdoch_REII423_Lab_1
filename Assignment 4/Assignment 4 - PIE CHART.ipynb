```python
#Import a csv file and plot a PIE CHART

import csv
import pandas as pd
from matplotlib import pyplot as plt

filename = "AirPassengers.csv"
f = open(filename,"r")
df = pd.read_csv(f)

sums_by_passengers = df.groupby('Date').sum()
var = 'AirPassengers'
sums_by_passengers[var].plot(kind='pie', fontsize=20)
plt.title('Air passengers per year', fontsize=25)
plt.savefig('PieChartAirPassengersByDate.jpg', dpi = 600)
plt.show()
```
