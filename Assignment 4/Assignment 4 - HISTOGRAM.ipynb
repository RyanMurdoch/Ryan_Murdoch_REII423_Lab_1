```python
#Import a csv file and plot a HISTOGRAM

import matplotlib.pyplot as plt
from numpy.random import normal
gaussian_numbers = normal(size=2000)
plt.hist(gaussian_numbers)
plt.title("Gaussian Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.hist(gaussian_numbers, bins=500)
plt.savefig('GaussianHIST.jpg', dpi = 1200)
plt.show()
```
