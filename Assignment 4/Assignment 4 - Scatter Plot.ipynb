```python
import pandas

import matplotlib.pyplot as plt

data = pandas.read_csv('brain_size.csv', sep=',', na_values='.')
gender_data = data.groupby('Gender')

figure1 = plt.scatter(data['Weight'], data['Height'])
plt.xlabel('Weight')
plt.ylabel('Height')
plt.title('Weight vs. Height')
plt.savefig('BrainScatter.jpg', dpi = 1200)
plt.show()
```
